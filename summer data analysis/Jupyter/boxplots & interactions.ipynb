```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
#import csvs
url_1 = 'rs_removed_part1.csv'
url_2 = 'rs_removed_part2.csv'
part_1 = pd.read_csv(url_1)
part_2 = pd.read_csv(url_2)
```

```python
#remove extra column
part_1 = part_1[["ID", "composer", "condition", "name", "response_value"]]
part_2 = part_2[["ID", "composer", "condition", "name", "response_value"]]
```

```python
#side by side boxplots for part1 coherence
coherence_frame_1 = part_1[part_1["name"] == "coherence"]

coh1_scrambled = coherence_frame_1[coherence_frame_1.condition == "scrambled"]
coh1_reversed = coherence_frame_1[coherence_frame_1.condition == "reversed"]
coh1_forward = coherence_frame_1[coherence_frame_1.condition == "forward"]
```

```python
sns.set(style="whitegrid", palette="muted")
coh1_plot = sns.pointplot(x="condition", y="response_value", hue="composer", hue_order=["bach", "chopin"], data=coherence_frame_1)
coh1_plot.set_xticklabels(coh1_plot.get_xticklabels(), rotation=45)
coh1_plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
