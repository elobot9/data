```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
#import csvs
url_1 = 'part_1.csv'
url_2 = 'part_2.csv'
part_1 = pd.read_csv(url_1)
part_2 = pd.read_csv(url_2)
```

```python
#remove 'Unnamed:0 column
part_1 = part_1[['ID', 'composer', 'condition', 'name', 'response_value']]
part_2 = part_2[['ID', 'composer', 'condition', 'name', 'response_value']]
```

```python
#label part 1 rows as part 1 and part 2 rows as part 2

part_1['part'] = 'part_1'
part_2['part'] = 'part_2'
```

```python
#concatenate the two dfs
full_exp = pd.concat([part_1, part_2])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
sns.set(style="whitegrid", palette="muted")
full_plot = sns.boxplot(x="name", y="response_value", hue="part", hue_order=["part_1", "part_2"], order=['affect', 'affect-reverse', 'coherence', 'coherence-reverse', 'dynamism', 'dynamism-reverse', 'expressiveness', 'expressiveness-reverse', 'strangeness', 'strangeness-reverse', 'desire-to-own', 'enjoyment'], data=full_exp)
full_plot.set_xticklabels(full_plot.get_xticklabels(), rotation=90)
full_plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(style="whitegrid", palette="muted")
full_plot = sns.boxplot(x="part", y="response_value", data=full_exp)
full_plot.set_xticklabels(full_plot.get_xticklabels(), rotation=90)
full_plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
