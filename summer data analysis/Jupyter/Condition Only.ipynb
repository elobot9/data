```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
#import csvs
url_1 = 'rs_removed_part1.csv'
url_2 = 'rs_removed_part2.csv'
part_1 = pd.read_csv(url_1)
part_2 = pd.read_csv(url_2)
```

```python
#remove extra column
part_1 = part_1[["ID", "composer", "condition", "name", "response_value"]]
part_2 = part_2[["ID", "composer", "condition", "name", "response_value"]]
```

```python
full = part_1.append(part_1)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python

sns.set(style="whitegrid", palette="muted")
full_plt = sns.pointplot(x="name", y="response_value", hue="condition", data=full, hue_order=['forward', 'reversed', 'scrambled'],order=["affect", "coherence", "desire-to-own", "dynamism", "enjoyment", "expressiveness", "strangeness"], dodge=True)
full_plt.set_xticklabels(full_plt.get_xticklabels(), rotation=90)
full_plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
full_plt.set_title("Ratings per condition for all questions")
full_plt.set(ylim=(0, 4))
```

```python
enjoyment = full[full["name"] == "enjoyment"]

sns.set(style="whitegrid", palette="muted")
enjoy_plot = sns.pointplot(x="condition", y="response_value",data=enjoyment, order=['forward', 'reversed', 'scrambled'], dodge=True)
enjoy_plot.set_xticklabels(enjoy_plot.get_xticklabels(), rotation=45)
enjoy_plot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
enjoy_plot.set_title("Enjoyment")
enjoy_plot.set(ylim=(0,4))
```
