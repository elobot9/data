```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pylab as pl
```

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('plot')
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x,y)
ax1.set_title('share y axis')
ax2.scatter(x, y)
```
